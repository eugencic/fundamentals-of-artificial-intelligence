## import required libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

## initialize data

```python
# Generate random data
np.random.seed(0)
X = np.random.rand(100, 1)
y = 2 * X + 1 + 0.1 * np.random.randn(100, 1)
```

## Split data into train and test

```python
num_samples = len(X)
split_ratio = 0.8

X_train, X_test = X[:int(num_samples * split_ratio)], X[int(num_samples * split_ratio):]
y_train, y_test = y[:int(num_samples * split_ratio)], y[int(num_samples * split_ratio):]
```

## Create the NN model

```python
# Create a Sequential model
model = tf.keras.Sequential()

# Add a single Dense layer for linear regression
model.add(tf.keras.layers.Dense(1, input_dim=1))


model.compile(loss='mean_squared_error', optimizer='sgd')
model
```

## Train the model

```python
history = model.fit(X_train, y_train, epochs=300, verbose=1)
```

```python
plt.plot(history.history['loss'], label="training loss")
plt.legend()
```

## Get predictions on the test set

```python
y_pred = model.predict(X_test)
```

## Plot the results on the test set

```python
plt.scatter(X_test, y_test, label='True Data')
plt.scatter(X_test, y_pred, label='Predictions')
plt.legend()
plt.show()
```

## Evaluate the model - compute the losses on train and test data

```python
mse_loss_train = model.evaluate(X_train, y_train)
mse_loss_test = model.evaluate(X_test, y_test)

print("MSE loss on train data:", mse_loss_train)
print("MSE loss on test data:", mse_loss_test)
```
